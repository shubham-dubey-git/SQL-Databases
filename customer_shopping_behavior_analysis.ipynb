##### **Replacing the Null Value Using Median for Each Category**

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

##### **Snake Casing All Columns**

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

#### **Create a New Column (age_group)**

```python
# Used qcut to split the 'age' column into 4 equal-sized quantile-based age groups
# q = 4  â†’ divides the data into 4 quantiles (25% each)

labels = ['Young Adult', 'Adult', 'Middle-Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)  
```

```python
df[['age', 'age_group']].head(10)
```

#### **Create Column Purchase Frequency Days**

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['puchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['puchase_frequency_days','frequency_of_purchases']].head(10)
```
